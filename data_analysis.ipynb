```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')   # read the csv fiel and convert to pandas dataframe
```

```python
specific = df.iloc[[2,3,6]]  #select students with their indices

name=specific['Name of student']
marks=specific['hrs']   # select category to compare ie 'hrs' can be changed to 'AEC'

get_rows=df.head(30) # df.head(no of rows) selects desired no of rows from to of data frame(df)



plt.xlabel('name')    
plt.ylabel('marks')

plt.bar(name,marks)  #plt.bar plotting function in pyplot
```

```python
# filters based on req criteron and gives list
print('AEC FILTER')
marks_filter_AEC=get_rows[get_rows["AEC"]>=90]     
print(marks_filter_AEC)
print('\n\n')


print('SS FILTER')
marks_filter_SS=get_rows[get_rows["SS"]>=90]
print(marks_filter_SS)
print('\n\n')



print('DSD FILTER')
marks_filter_DSD=get_rows[get_rows["DSD"]>=90]
print(marks_filter_DSD)
print('\n\n')


print('NT FILTER')
marks_filter_NT=get_rows[get_rows["NT"]>=90]
print(marks_filter_NT)

print('STUDY hrs')
marks_filter_hrs=get_rows[get_rows["hrs"]>=20]
print(marks_filter_hrs)




```

```python
import numpy


temp=df.iloc[[2]]    # change value of '2' to desired index from csv 
print(temp)

arr = temp.to_numpy()   # use numpy to convert to 1d array

hah=numpy.delete(arr, [0,6,7,8]) #deleting characters to get array with only floats




labels=['AEC','SS','DSD','MATH','NT'] #labels for pie chart




myexplode = [0, 0.2, 0, 0,0] # change to 0.2 to explode corresponding label

plt.pie(x=hah, labels = labels, explode = myexplode, shadow = True) # 'x' are the marks and 'labels' is the name of subject 
```

```python





std_1=df.iloc[[2]]   #select student 1
std_2=df.iloc[[3]]   #select student 2

np1=std_1.to_numpy()   #convert to numpy array to get 1d list
np2=std_2.to_numpy()

temp1=numpy.delete(np1,[0,6,7,8])  #delete unecessary data points ie avg, std hrs etc

temp2=numpy.delete(np2,[0,6,7,8])


subjects=['AEC','SS','DSD','MATH','NT'] # create labels
stduno=temp1   # pass temp vars 
stddos=temp2


#create new dataframe from above vars
df_days_calories = pd.DataFrame(
    { 'Subject' : subjects ,  
     'student1': stduno ,  
     'student2': stddos}) 



# print student names
print("student1=",np1[0])
print("student2=",np2[0])
ax = plt.gca() 
  
#use plot() method on the dataframe 
#create line graph to comare marks
plt.title("Marks Comparison")
df_days_calories .plot( x = 'Subject' , y = 'student1', ax = ax ) 
df_days_calories .plot( x = 'Subject' , y = 'student2' , ax = ax)
```
